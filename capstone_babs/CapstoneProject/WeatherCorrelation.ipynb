```python
%matplotlib inline
import pandas as pd


topTripCountsFile = './data/201402_TopStation_Weekday_12Count.csv'

topTripCounts = pd.DataFrame.from_csv(topTripCountsFile)
topTripCounts=topTripCounts.reset_index()
topTripCounts
```

```python

sfoTempFile = './data/SFO_temperatures.csv'

sfoTemp = pd.DataFrame.from_csv(sfoTempFile)
sfoTemp=sfoTemp.reset_index()
# sfoTemp=sfoTemp.groupby('Time').mean().reset_index()
sfoTemp


tempDict={}
for index, row in sfoTemp.iterrows():
        tempDict[row['Time']] = (row['Temp(F)'],row['Humidity(%)'],row['WindSpeed(mph)'])
        
# tempDict
```

```python
topTripCounts['temp']=topTripCounts['starthour'].map(lambda x: tempDict[x][0])
topTripCounts['humid']=topTripCounts['starthour'].map(lambda x: tempDict[x][1])
topTripCounts['windspeed']=topTripCounts['starthour'].map(lambda x: tempDict[x][2])
# topTripCounts['starthour']
topTripCounts.head()
# tripStationHourCounts['hour']=tripStationHourCounts['starthour'].map(lambda x: x.hour)
```

```python
import matplotlib.pylab as plt

plt.scatter(topTripCounts.temp, topTripCounts['50'])
plt.show()
plt.scatter(topTripCounts.humid, topTripCounts['50'])
plt.show()
plt.scatter(topTripCounts.windspeed, topTripCounts['50'])
```
