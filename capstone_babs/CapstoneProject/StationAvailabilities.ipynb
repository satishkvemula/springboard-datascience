```python
#Get Station Dict
import pandas as pd
stationData = pd.DataFrame.from_csv('./data/201402_station_data.csv')
# help(sampleTripMilestones)

stationDict={}
for index, row in stationData.reset_index().iterrows():
    stationDict[row['station_id']] = row['dockcount']
```

```python

def availableDocks(row):
    return stationDict[row['Station']] - row['Change']

stationChangesByDay = pd.DataFrame.from_csv('./extractedData/201402_StationChangesByDay.csv')
stationChangesByDay['Available'] = stationChangesByDay.apply(availableDocks,axis=1)

stationChangesByDay.to_csv('./extractedData/201402_StationChangesByDay_WithAvailable.csv',index=False)
runningLowInDay=stationChangesByDay[stationChangesByDay['Available']< 5]
runningLowInDay
```

```python

stationChangesByHour = pd.DataFrame.from_csv('./extractedData/201402_StationChangesByHour.csv')
stationChangesByHour['Available'] = stationChangesByHour.apply(availableDocks,axis=1)

stationChangesByHour.to_csv('./extractedData/201402_StationChangesByHour_WithAvailable.csv',index=False)
runningLowInHour=stationChangesByHour[stationChangesByHour['Available']< 5]
runningLowInHour
```
