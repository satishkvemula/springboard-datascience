```python
%matplotlib inline
import pandas as pd


tripsDataFile = '../201402_Trips_WithBicycleDistanceSpeed.csv'

trips = pd.DataFrame.from_csv(tripsDataFile, index_col='Trip ID', encoding='utf-8')
trips = trips[(trips['Duration'] <= 5400) & (trips['Duration'] > 60)]

# startStationTripCounts  = trips[['Start Station']].groupby(['Start Station']).agg(['count'])

# startStationTripCounts=pd.DataFrame(trips.groupby(['startStationId']).size().rename('counts'))
# startStationTripCounts = startStationTripCounts.sort(['counts'], ascending=[0])
# startStationTripCounts

# .sort_values(by='Duration', ascending=False).head(n=15)

# Group by zones, and do a mean of the count, do a mean of the duration, 
```

```python
# import pandas as pd
# stationData = pd.DataFrame.from_csv('../201402_station_data.csv')
# # help(sampleTripMilestones)

# stationDict={}
# for index, row in stationData.reset_index().iterrows():
#     stationDict[row['name']] = row['station_id']

zoneCount={}
zoneCount['SJC']= 16
zoneCount['RDWD']=7
zoneCount['MVIEW']=7
zoneCount['PALTO']=5
zoneCount['SFO']=35

def groupByZone(stationId):
    stationId=int(stationId)
    zone = ''
    if ( 2 <= stationId <=16) :
        zone = "SJC"
    elif (21 <= stationId <= 26) :
        zone = "RDWD"
    elif (27 <= stationId <= 33) :
        zone = "MVIEW"
    elif (34 <= stationId <= 38) :
        zone = "PALTO"
    elif (39 <= stationId <= 77) :
        zone = "SFO"
    elif (stationId == 80) :
        zone = "SJC"
    elif (stationId == 83) :
        zone = zone = "RDWD"
    elif (stationId == 82) :
        zone = "SFO"
    else:
        zone = "UNK"
    return zone

trips['stationZone']=trips['startStationId'].map(groupByZone)
# trips['stationZone','startStationId']
tripStationZones = trips['stationZone'].reset_index()
tripStationZones
# zoneStationCounts = trips['stationZone','stationId'].groupby(['stationZone', 'stationId']).agg({'tripCount': 'count'})
zoneStationCounts = tripStationZones.groupby(['stationZone'])['Trip ID'].count().reset_index(name="tripCount")
zoneStationCounts['zoneStationCount'] = zoneStationCounts['stationZone'].map(lambda zone: zoneCount[zone])
zoneStationCounts['tripsPerStation']=zoneStationCounts['tripCount']/zoneStationCounts['zoneStationCount']


zoneStationCounts
```
