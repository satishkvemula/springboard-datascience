```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
tripDurationDataFile = '201402_Trips_WithBicycleDistanceSpeed.csv'
```

```python
df = pd.read_csv(tripDurationDataFile)
```

```python
reasonableTrips=df[(df['Duration'] <= 2400) & (df['Duration'] > 120)]
```

```python
# reasonableTrips['Start Date'] = pd.to_datetime(reasonableTrips['Start Date'])  
reasonableTrips['starttime'] = reasonableTrips['Start Date'].apply(lambda x: x.weekday())
```

```python
weekdayTrips = reasonableTrips[(reasonableTrips['starttime'] < 5 )]
weekendTrips = reasonableTrips[(reasonableTrips['starttime'] >4 )]
```

```python
reasonableTrips.boxplot(column='Duration', by='startStationId', figsize=(20,10)) 
weekdayTrips.boxplot(column='Duration', by='startStationId', figsize=(20,10)) 
weekendTrips.boxplot(column='Duration', by='startStationId', figsize=(20,10))
```

```python

print(len(reasonableTrips.index))
print(len(weekdayTrips.index))
print(len(weekendTrips.index))
```

```python
mondayTrips = weekdayTrips[(weekdayTrips['starttime'] ==0 )]
tuesdayTrips = weekdayTrips[(weekdayTrips['starttime'] ==1 )]
wednesTrips = weekdayTrips[(weekdayTrips['starttime'] ==2 )]
thursdayTrips = weekdayTrips[(weekdayTrips['starttime'] ==3 )]
fridayTrips = weekdayTrips[(weekdayTrips['starttime'] ==4 )]
```
